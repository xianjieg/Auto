```python
from IPython.display import display, HTML
import pandas as pd
import sqlite3
from sqlite3 import Error

def create_connection(db_file, delete_db=False):
    import os
    if delete_db and os.path.exists(db_file):
        os.remove(db_file)

    conn = None
    try:
        conn = sqlite3.connect(db_file)
        conn.execute("PRAGMA foreign_keys = 1")
    except Error as e:
        print(e)

    return conn


def create_table(conn, create_table_sql, drop_table_name=None):
    
    if drop_table_name: # You can optionally pass drop_table_name to drop the table. 
        try:
            c = conn.cursor()
            c.execute("""DROP TABLE IF EXISTS %s""" % (drop_table_name))
        except Error as e:
            print(e)
    
    try:
        c = conn.cursor()
        c.execute(create_table_sql)
    except Error as e:
        print(e)
        
def execute_sql_statement(sql_statement, conn):
    cur = conn.cursor()
    cur.execute(sql_statement)

    rows = cur.fetchall()

    return rows

```

```python
def create_make_table(data_filename):

     
    normalized_database_filename = 'Auto.db'
    normalized_conn=create_connection(normalized_database_filename, True)
    
    
    create_table_make_sql="""CREATE TABLE [make] (
        [make] Text not null primary key
        
        
    ); """
        
   
    insert_Make=pd.read_csv(data_filename,sep=",")
    #display(insert_Make)
    make_list=[]
    make_list=insert_Make["make"].unique()
    #make_list=sorted(make_list)
    #print(make_list)
    
    make_list=[(ele,)for ele in make_list]
    
        
    def insert_make(conn, values):
        sql=''' INSERT INTO make( make) VALUES(?) '''
        cur = conn.cursor()
        cur.executemany(sql, values)
        #return cur.lastrowid
    
    with normalized_conn:
        create_table(normalized_conn, create_table_make_sql,"make")
        #print(make_tuple)
        insert_make(normalized_conn, make_list)
        
data_filename = 'auto .csv'       
create_make_table(data_filename)        
sql="select * from make;"
normalized_database_filename = 'Auto.db'
conn = create_connection(normalized_database_filename)
df = pd.read_sql_query(sql,conn)
display(df)
```

```python
def create_car_table(data_filename):

     
    normalized_database_filename = 'Auto.db'
    normalized_conn=create_connection(normalized_database_filename)
    
    
    create_table_car_sql="""CREATE TABLE [car] (
        [carID] Integer not null primary key,
        
        [make] integer not null,
        [fueltype] integer not null,
        [numofdoors] integer not null,
        [bodystyle] integer not null,
        [drivewheels] integer not null,
        [numofcylinders] integer not null,
        
        [enginesize] integer not null,      
        [horsepower] integer not null,        
        [peakrpm] integer not null,      
        [citympg] integer not null,        
        [highwaympg] integer not null,        
        [price] integer not null,

        
        FOREIGN KEY (make) REFERENCES make (make),
        FOREIGN KEY (fueltype) REFERENCES fueltype (fueltype),
        FOREIGN KEY (numofdoors) REFERENCES numofdoors (numofdoors),
        FOREIGN KEY (bodystyle) REFERENCES bodystyle (bodystyle),
        FOREIGN KEY (drivewheels) REFERENCES drivewheels (drivewheels),
        FOREIGN KEY (numofcylinders) REFERENCES numofcylinders (numofcylinders)
        
        
        
        
        

    ); """
        
   
    data=pd.read_csv(data_filename,sep=",")
    #display(data)
    
    
    
    my_list=[]
    for row in data.values.tolist():
        
        carID=row[0]
        make=row[1]
        fueltype=row[2]
        numofdoors=row[3]
        bodystyle=row[4]
        drivewheels=row[5]
        numofcylinders=row[6]
        
        #print(list(row))
        enginesize=row[7]
        #print(enginesize)
        horsepower=row[8]
        peakrpm=row[9]
        citympg=row[10]
        highwaympg=row[11]
        price=row[12]
        
        
        
        my_list.append((carID,make, fueltype, numofdoors, bodystyle, drivewheels, numofcylinders, 
                          enginesize, horsepower, peakrpm, citympg, highwaympg, price))
    
  
        
    def insert_car(conn, values):
        sql=''' INSERT INTO car(carID,make, fueltype, numofdoors, bodystyle, drivewheels, numofcylinders, 
                          enginesize, horsepower, peakrpm, citympg, highwaympg, price) VALUES(?,?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?) '''
       
        cur = conn.cursor()
        cur.executemany(sql, values)
        return cur.lastrowid
    
    with normalized_conn:
        create_table(normalized_conn, create_table_car_sql,"car")
        insert_car(normalized_conn, my_list)
        normalized_conn.commit;
        
data_filename = 'auto .csv'       
create_car_table(data_filename)        
sql="select * from car;"
normalized_database_filename = 'Auto.db'
conn = create_connection(normalized_database_filename)
df = pd.read_sql_query(sql,conn)
display(df)
```
